```python
# Import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#loading restaurants CSV
restaurants_path ="restaurants.csv"
restaurants_df = pd.read_csv(restaurants_path)
restaurants_df.head()
```

```python
clean_restaurants_df = restaurants_df.drop(['index'], axis=1)
clean_restaurants_df
```

```python

#Reset index name to load into SQL database
clean_restaurants_df.index.name = "Index"
clean_restaurants_df
```

```python
#connect engine
###note to group, update local host location and name of database to reflect what is on your system
connection_string = "postgres:postgres@localhost:5432/final_project"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#checking tables names 
engine.table_names()
```

```python
#loading data to postgresSQL
clean_restaurants_df.to_sql(name='restaurants', con=engine, if_exists='append', index=True)
```

```python
#loading schools CSV
schools_path ="schools.csv"
schools_df = pd.read_csv(schools_path)
schools_df.head()
```

```python
clean_schools_df = schools_df.drop(['index'], axis=1)
clean_schools_df
```

```python
#Reset index name to load into SQL database
clean_schools_df.index.name = "Index"
clean_schools_df
```

```python
final_schools_df = clean_schools_df.rename(columns={"postcode": "zipcode", "Borough": "borough", "Latitude": "latitude", "Longitude": "longitude", "Community Board": "community_board", "Council District": "council_district", "Census Tract": "census_tract", "NTA":"nta"})


```

```python
#loading data to postgresSQL
final_schools_df.to_sql(name='schools', con=engine, if_exists='append', index=True)
```
